```python
# Plot the number of employees in each department as a bar chart.
numberOfEmploees = data["Department"].value_counts().sort_values(ascending=False).head(4)
plt.bar(numberOfEmploees.index, numberOfEmploees.values, color='skyblue')
plt.show()
```

```python
# Create a pie chart showing the gender distribution of employees.
gender=data["Sex"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    gender.values,
    labels=gender.index,       
    autopct="%1.1f%%",        
    startangle=90,            
    colors=['skyblue', 'lightpink']  
)
plt.title("Gender Distribution of Employees")
plt.show()
```

```python
# Make a histogram showing the age distribution of employees.

plt.figure(figsize=(10,6))
plt.hist(data["Age"], bins=10, color='skyblue', edgecolor='black')  
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Age Distribution of Employees")
plt.grid(axis='y', alpha=0.75)  
plt.show()
```

```python
# Plot a bar chart of the average salary per department.
plt.figure(figsize=(10,6))
averagesalary = data.groupby("Department")["Salary"].mean()
plt.bar(averagesalary.index,averagesalary.values,color="lightpink",edgecolor="Black")
plt.title("Average Salary per Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.grid(axis="y",alpha=0.5)
plt.show()
```

```python
# Show a line chart of employee hires by year (DateofHire).
data["DateofHire"] = pd.to_datetime(data["DateofHire"])
hire_per_year = data["DateofHire"].dt.year.value_counts().sort_index()
hire_per_year
plt.figure(figsize=(10,6))
plt.plot(hire_per_year.index,hire_per_year.values,marker="o",color="blue")
plt.xlabel("Year")
plt.ylabel("Number of Hires")
plt.title("Employee Hires by Year")
plt.grid(True)
plt.show()
```

```python
# Create a boxplot showing the salary distribution across departments.

plt.figure(figsize=(12,6))
data.boxplot(column='Salary', by='Department', grid=False)
plt.title("Salary Distribution by Department")
plt.suptitle("")  # removes the default 'Boxplot grouped by group_by_column_name' title
plt.xlabel("Department")
plt.ylabel("Salary")
plt.xticks(rotation=45)  # rotate x labels for better readability
plt.show()
```

```python
# Draw a bar chart showing the count of each Marital Status category.
marital_map = {0: "Single", 1: "Married", 2: "Divorced"}
d=data["MaritalStatusID"].map(marital_map).value_counts()
plt.figure(figsize=(9,5))
plt.bar(d.index,d.values,color="green",edgecolor="Red")
plt.grid(axis="y",color="Black")
plt.title("count of each Marital Status category")
plt.xlabel("Marital Status category")
plt.ylabel("count of each Marital Status")
plt.show()

```
